```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/1.csv")
data.head()
```

```python
aggregated_data = data.groupby(["threads_per_block", "number_of_elements"])["time"].aggregate(["mean", "std"])
aggregated_data
```

```python
data = aggregated_data.reset_index()
```

```python
sns.lineplot(data=data, x="number_of_elements", y="time", hue="")
```
